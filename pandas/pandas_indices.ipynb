 Hay una cierta equivalencia entre DataFrames y tablas de una base de datos.
 10 km
 Los √≠ndices pueden ser muy √∫tiles cuando haya que seleccionar datos en un gran dataframe, o si hay que unir dos de
 ellos.
 Para crear un √≠ndice usaremos la funci√≥n set_index:


```python
df4=pd.DataFrame({'Estacion':['Los Platos','Atima','Bomberos_Tegucigalpa'],
                  'Date':[datetime(2023,8,27,12),
                          datetime(2023,8,27,12),
                          datetime(2023,8,27,12)],
                  'PCP 1h':[0.0,0.0,0.0],
                  'PCP 6h':[7.6,0.1,0.0],
                  'PCP 12h':[11.7,0.1,7.3],
                  'PCP 24h':[11.7,19.2,18.8]
                  })
df4
```

```python
df4.set_index('Date',inplace=True)
df4
```

 El √≠ndice tambi√©n se puede definir simult√°neamente al crear el dataframe:

```python
df5=pd.DataFrame({
    'Estacion':['Los Platos','Atima','Bomberos-Tegucigalpa'],
    'PCP 1h':[0.0,0.0,0.0],
    'PCP 6h':[7.6,0.1,0.0],
    'PCP 12h':[11.7,0.1,7.3],
    'PCP 24h':[11.7,19.2,18.8]},
    index=[datetime(2023,8,27,12),
           datetime(2023,8,27,12),
           datetime(2023,8,27,12)])

df5.index.name='Date'
df5
```

Para eliminar el √≠ndice:

```python
df5.reset_index(inplace=True)

df5
```

 Observar que muchas funciones, como set_index, devuelven un dataframe como salida:

```python
df5.set_index('Date',inplace=True)
df5
```

 Es posible crear multi√≠ndices, √≠ndices de m√°s de un campo:

```python
df5.reset_index(inplace=True)
df5
```

```python
df5.set_index(['Date','Estacion'],inplace=True)
df5
```
