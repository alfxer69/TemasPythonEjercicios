 Los índices pueden ser muy útiles cuando haya que seleccionar datos en un gran dataframe, o si hay que unir dos de
 ellos.
 Para crear un índice usaremos la función set_index:


```python
df4=pd.DataFrame({'Estacion':['Los Platos','Atima','Bomberos_Tegucigalpa'],
                  'Date':[datetime(2023,8,27,12),
                          datetime(2023,8,27,12),
                          datetime(2023,8,27,12)],
                  'PCP 1h':[0.0,0.0,0.0],
                  'PCP 6h':[7.6,0.1,0.0],
                  'PCP 12h':[11.7,0.1,7.3],
                  'PCP 24h':[11.7,19.2,18.8]
                  })
df4
```

```python
df4.set_index('Date',inplace=True)
df4
```

 El índice también se puede definir simultáneamente al crear el dataframe:

```python
df5=pd.DataFrame({
    'Estacion':['Los Platos','Atima','Bomberos-Tegucigalpa'],
    'PCP 1h':[0.0,0.0,0.0],
    'PCP 6h':[7.6,0.1,0.0],
    'PCP 12h':[11.7,0.1,7.3],
    'PCP 24h':[11.7,19.2,18.8]},
    index=[datetime(2023,8,27,12),
           datetime(2023,8,27,12),
           datetime(2023,8,27,12)])

df5.index.name='Date'
df5
```

Para eliminar el índice:

```python
df5.reset_index(inplace=True)

df5
```

 Observar que muchas funciones, como set_index, devuelven un dataframe como salida:

```python
df5.set_index('Date',inplace=True)
df5
```

 Es posible crear multiíndices, índices de más de un campo:

##Visualización de las características de un DataFrame

 Para manejar adecuadamente un dataframe debemos conocer sus características: qué columnas tiene, qué tipo de datos
 almacena, etc. Hay varias propiedades y métodos que nos dan esa información.

```python
# Leer el archivo como texto
with open("/content/registro_temperatura365d_smn.txt", "r",encoding="iso-8859-1") as file:  # Cambia el nombre si es necesario
    lines = file.readlines()

# Procesar cada línea
data = []
for line in lines:
    line = line.strip()  # Eliminar espacios y saltos de línea

    # Saltar líneas vacías o con guiones (---)
    if not line or "---" in line:
        continue

    # Dividir la línea en componentes (FECHA, TMAX, TMIN, NOMBRE)
    parts = line.split()

    # Extraer FECHA (primeros 8 dígitos)
    fecha = parts[0]

    # Extraer TMAX (puede ser numérico o 'NA')
    tmax = parts[1] if len(parts) > 1 else None

    # Extraer TMIN (puede ser numérico o negativo)
    tmin = parts[2] if len(parts) > 2 else None

    # El resto es el NOMBRE (puede tener espacios)
    nombre = " ".join(parts[3:]) if len(parts) > 3 else None

    # Guardar en la lista de datos
    data.append([fecha, tmax, tmin, nombre])
```

```python
#covertir a DataFrame
df=pd.DataFrame(data,columns=['FECHA','TMAX','TMIN','NOMBRE'])
df
```

```python
df_final=df.dropna()
df_final
```

```python
df_final=df_final.set_index('FECHA')
df_final.head(10)
```
