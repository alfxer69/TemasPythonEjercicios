##Visualizaci√≥n de las caracter√≠sticas de un DataFrame

 Para manejar adecuadamente un dataframe debemos conocer sus caracter√≠sticas: qu√© columnas tiene, qu√© tipo de datos
 almacena, etc. Hay varias propiedades y m√©todos que nos dan esa informaci√≥n.

```python
# Leer el archivo como texto
with open("/content/registro_temperatura365d_smn.txt", "r",encoding="iso-8859-1") as file:  # Cambia el nombre si es necesario
    lines = file.readlines()

# Procesar cada l√≠nea
data = []
for line in lines:
    line = line.strip()  # Eliminar espacios y saltos de l√≠nea

    # Saltar l√≠neas vac√≠as o con guiones (---)
    if not line or "---" in line:
        continue

    # Dividir la l√≠nea en componentes (FECHA, TMAX, TMIN, NOMBRE)
    parts = line.split()

    # Extraer FECHA (primeros 8 d√≠gitos)
    fecha = parts[0]

    # Extraer TMAX (puede ser num√©rico o 'NA')
    tmax = parts[1] if len(parts) > 1 else None

    # Extraer TMIN (puede ser num√©rico o negativo)
    tmin = parts[2] if len(parts) > 2 else None

    # El resto es el NOMBRE (puede tener espacios)
    nombre = " ".join(parts[3:]) if len(parts) > 3 else None

    # Guardar en la lista de datos
    data.append([fecha, tmax, tmin, nombre])
```

```python
#covertir a DataFrame
df=pd.DataFrame(data,columns=['FECHA','TMAX','TMIN','NOMBRE'])
df
```

De todas formas, en un programa es mejor utilizar los m√©todos loc e iloc que est√°n m√°s optimizados. Adem√°s estos
m√©todos nos evitan problemas relacionados con vistas y copias.
loc permite seleccionar subconjuntos de nuestra estructura (serie o dataframe), seg√∫n etiquetas. Para seleccionar
columnas:


```python
df_final[['TMAX','TMIN']]
```

```python
df_final.loc[:,('TMAX','TMIN')]
df_final.head(5)
```

y para seleccionar filas:


```python
df_final.loc['2024-05-17',:]
df_final.head(5)
```

o ambos a la vez:

```python
df_final.loc['2024-05-17':'2025-05-16',('TMAX','TMIN')]
df_final.head(5)
```

Nota: observar que en la selecci√≥n por etiquetas, tanto el elemento inicial como el final se incluyen (al contrario de lo
que ocurre en una lista)
iloc selecciona subconjuntos seg√∫n su posici√≥n. De forma equivalente a los ejemplos anteriores:

```python
df_final.iloc[4]
```

o tambi√©n:

```python
df_final.iloc[5:]
```

```python
df_final.iloc[1,1]
```

¬øCu√°ndo usar cada uno?
Usa loc cuando trabajes con nombres de filas/columnas (etiquetas, fechas, strings).

Usa iloc cuando necesites acceder por posici√≥n num√©rica (como en arrays de NumPy).

Si tu DataFrame tiene un √≠ndice num√©rico, loc seguir√° usando las etiquetas, mientras que iloc siempre usar√° la posici√≥n.

```python
#Ejemplo numerico
df_number=pd.DataFrame({
    'A':[20,30,40]
},index=[5,6,7])
```

```python
df_number.loc[5:7]
```

```python
df_number.iloc[0:2]
```
