##Visualización de las características de un DataFrame

 Para manejar adecuadamente un dataframe debemos conocer sus características: qué columnas tiene, qué tipo de datos
 almacena, etc. Hay varias propiedades y métodos que nos dan esa información.

```python
# Leer el archivo como texto
with open("/content/registro_temperatura365d_smn.txt", "r",encoding="iso-8859-1") as file:  # Cambia el nombre si es necesario
    lines = file.readlines()

# Procesar cada línea
data = []
for line in lines:
    line = line.strip()  # Eliminar espacios y saltos de línea

    # Saltar líneas vacías o con guiones (---)
    if not line or "---" in line:
        continue

    # Dividir la línea en componentes (FECHA, TMAX, TMIN, NOMBRE)
    parts = line.split()

    # Extraer FECHA (primeros 8 dígitos)
    fecha = parts[0]

    # Extraer TMAX (puede ser numérico o 'NA')
    tmax = parts[1] if len(parts) > 1 else None

    # Extraer TMIN (puede ser numérico o negativo)
    tmin = parts[2] if len(parts) > 2 else None

    # El resto es el NOMBRE (puede tener espacios)
    nombre = " ".join(parts[3:]) if len(parts) > 3 else None

    # Guardar en la lista de datos
    data.append([fecha, tmax, tmin, nombre])
```

```python
#covertir a DataFrame
df=pd.DataFrame(data,columns=['FECHA','TMAX','TMIN','NOMBRE'])
df
```

De todas formas, en un programa es mejor utilizar los métodos loc e iloc que están más optimizados. Además estos
métodos nos evitan problemas relacionados con vistas y copias.
loc permite seleccionar subconjuntos de nuestra estructura (serie o dataframe), según etiquetas. Para seleccionar
columnas:


```python
df[['TMAX','TMIN']]
```

```python
df.loc[:,('TMAX','TMIN')]
```

y para seleccionar filas:


```python
df.loc['2024-05-17',:]
```

o ambos a la vez:

```python
df.loc['2024-05-17':'2025-05-16',('TMAX','TMIN')]
```

Nota: observar que en la selección por etiquetas, tanto el elemento inicial como el final se incluyen (al contrario de lo
que ocurre en una lista)
iloc selecciona subconjuntos según su posición. De forma equivalente a los ejemplos anteriores:

```python
df.iloc[4]
```

o también:

```python
df.iloc[5:]
```
